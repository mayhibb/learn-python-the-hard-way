```python
ten_things = 'Apples Oranges Crows Telephone Lingt Sugar'

print("What there's not 10 things in that list, let's fix that.")

stuff = ten_things.split(' ')  # split（） 用符号分割字符串
more_stuff = ["Day", "Night", "Song", "Frisbee", "Corn", "Banana", "Girl", "Boy"]

while len(stuff) != 10:    # len 代表字符串的数量
    next_one = more_stuff.pop()  # pop（）默认为 index=-1，删除最后一个列表值。
    print("Adding:", next_one)
    stuff.append(next_one)
    print("There's %d items now." % len(stuff))
    
print("There we go:", stuff)

print("Let's do some things with stuff.")

print(stuff[1]) # 第二个
print(stuff[-1])  # 最后一个
print(stuff.pop()) 
print('/'.join(stuff)) # join（）链接字符串数组。
print("_".join(stuff))
print('#'.join(stuff[3:5]))# 链接基数的3和5，序数的4和6.
```
